```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt
```

```python
sns.distplot(iris["sepal_length"])
```

```python
iris.plot(kind = 'scatter',x = 'sepal_length',y = 'sepal_width')
```

```python
plt.scatter(iris["sepal_length"],iris["sepal_width"])
```

```python
sns.scatterplot(x = "sepal_length",y = "sepal_width",data = iris)
```

```python
sns.lineplot(x ="sepal_length",y = "sepal_width",data = iris)
```

```python
sns.barplot(x ="sepal_width",y = "petal_width",data = iris)
```

```python
sns.boxplot(x ="sepal_length",y = "sepal_width",data = iris)
```

```python
sns.catplot(x ="sepal_length",y = "petal_length",data = iris)
```

```python
sns.stripplot(x ="species",y = "sepal_length",data = iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot = True)
```
